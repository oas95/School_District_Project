```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combining the both csv files into a single dataset  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Finding unique school names for total amount of schools
total_schools = len(school_data_complete.school_name.unique())

#Total num of students
total_students = school_data_complete.student_name.count()

#Total budget
total_budget = sum(school_data_complete.budget.unique())

#Average Math Score
avgmath_score = school_data_complete.math_score.mean()

#Average Reading Score
avgreading_score = school_data_complete.reading_score.mean()

# % of students passing math
passing_math_per = (school_data_complete[school_data_complete['math_score'] >= 70].math_score.count()) / total_students

# % of students passing reading
passing_reading_per = (school_data_complete[school_data_complete['reading_score'] >= 70].reading_score.count()) / total_students

#Passing rate
overall_pass_rate = (passing_math_per + passing_reading_per)/2
```

```python
#storing district summary in df(DataFrame)
district_sum = pd.DataFrame({"Total Schools":[total_schools],
                                 "Total Students":[total_students],
                                 "Total Budget":[total_budget],
                                 "Average Math Score":[avgmath_score],
                                 "Average Reading Score":[avgreading_score],
                                 "% Passing Math":[passing_math_per*100],
                                 "% Passing Reading":[passing_reading_per*100],
                                 "% Overall Passing Rate":[overall_pass_rate*100]})
# Formating Total Students and Total Budget
district_sum['Total Students'] = district_sum.apply(lambda x: "{:,.0f}".format(x['Total Students']), axis=1)
district_sum['Total Budget'] = district_sum.apply(lambda x: "${:,.2f}".format(x['Total Budget']), axis=1)
#Formating for decimal places 2 sig figures
district_sum['Average Math Score'] = district_sum.apply(lambda x: "{:,.2f}".format(x['Average Math Score']), axis=1)
district_sum['Average Reading Score'] = district_sum.apply(lambda x: "{:,.2f}".format(x['Average Reading Score']), axis=1)
district_sum['% Passing Math'] = district_sum.apply(lambda x: "{:,.2f}%".format(x['% Passing Math']), axis=1)
district_sum['% Passing Reading'] = district_sum.apply(lambda x: "{:,.2f}%".format(x['% Passing Reading']), axis=1)
district_sum['% Overall Passing Rate'] = district_sum.apply(lambda x: "{:,.2f}%".format(x['% Overall Passing Rate']), axis=1)
# Summary df
district_sum
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Summary
school_summary = school_data_complete.groupby(['school_name'])

#School Names
school_names = school_data_complete.school_name.sort_values().unique()

#School Types
school_types = school_data.sort_values(by="school_name").type

#Total Students
school_total_students = list(school_summary.student_name.count())

#Total Budget
school_budget = list(school_summary.budget.mean())

#Budget per Student (i and j presented in this was are representing school_budget and total_students for this specific equation)
school_per_student_budget = [i/j for i,j in zip(school_budget,school_total_students)]

#Avergae math score
school_avg_math_score = list(school_summary.math_score.mean())

#Average reading Score
school_avg_reading_score = list(school_summary.reading_score.mean())

# Calculating passing percentages
# % passing in math
school_summary = school_data_complete[school_data_complete['math_score'] >= 70].groupby(['school_name'])
school_per_passing_math = [(i/j)*100 for i,j in zip(school_summary.math_score.count(),school_total_students)]

#% passing in reading 
school_summary = school_data_complete[school_data_complete['reading_score'] >= 70].groupby(['school_name'])
school_per_passing_reading = [(i/j)*100 for i,j in zip(school_summary.reading_score.count(),school_total_students)]

# passing overall in both math and reading
school_overall_passing = [(i+j)/2 for i,j in zip(school_per_passing_math,school_per_passing_reading)]


# Creating df for School summary
school_sumdf = pd.DataFrame({"School Names":school_names,
                                  "School Type":school_types,
                                  "Total Students":school_total_students,
                                  "Total School Budget":school_budget,
                                  "Per Student Budget":school_per_student_budget,
                                  "Average Math Score":school_avg_math_score,
                                  "Average Reading Score":school_avg_reading_score,
                                  "% Passing Math":school_per_passing_math,
                                  "% Passing Reading":school_per_passing_reading,
                                  "Overall Passing Rate":school_overall_passing})

school_sumdf = school_sumdf.reset_index(drop=True)

#displaying school summary
school_sumdf
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Use the code provided to separate the data by grade and mean calculation
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")].groupby(["school_name"])['math_score'].mean()
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")].groupby(["school_name"])['math_score'].mean()
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")].groupby(["school_name"])['math_score'].mean()
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")].groupby(["school_name"])['math_score'].mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({"9th":ninth_graders, 
                                     "10th":tenth_graders, 
                                     "11th":eleventh_graders, 
                                     "12th":twelfth_graders })

# Minor data wrangling
math_scores_by_grade.index.name = None

#Formating for decimal places 2 sig figures
math_scores_by_grade['9th'] = math_scores_by_grade.apply(lambda x: "{:,.2f}".format(x['9th']), axis=1)
math_scores_by_grade['10th'] = math_scores_by_grade.apply(lambda x: "{:,.2f}".format(x['10th']), axis=1)
math_scores_by_grade['11th'] = math_scores_by_grade.apply(lambda x: "{:,.2f}".format(x['11th']), axis=1)
math_scores_by_grade['12th'] = math_scores_by_grade.apply(lambda x: "{:,.2f}".format(x['12th']), axis=1)

# Display the DataFrame
math_scores_by_grade
```

```python
# Establish the bins labels = how much of the budget per student per each school or spending per student reletive to each school
spending_bins = [0, 585, 630, 645, 680]
spending_range = ["<$585", "$585-630", "$630-645", "$645-680"]

#creating new df for scores by spending
school_spending_df = school_sumdf[["School Names",
                                    "Average Math Score",
                                    "Average Reading Score",
                                    "% Passing Math",
                                    "% Passing Reading",
                                    "Overall Passing Rate"]]
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_sumdf["Per Student Budget"], spending_bins, labels=spending_range)
school_spending_df = school_spending_df.groupby(["Spending Ranges (Per Student)"])
school_spending_df.mean()
```
